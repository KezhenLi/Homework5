```
import urllib.request
from numpy.fft import rfft, irfft
from numpy import loadtxt, zeros, copy
from matplotlib.pylab import plot, show
```

```
#(a)
url='http://www-personal.umich.edu/~mejn/cp/data/dow.txt'
with urllib.request.urlopen(url) as response:
     html = response.read()
data=html.decode().split("\n")
dow = loadtxt(data, float)
plot(dow)
show()
#(b)
dowFourier = rfft(dow)
#(c)
N = len(dowFourier)
first10percent = zeros(N, float)
first10percent[0 : int(N / 10)] = copy(dowFourier[0 : int(N / 10)])
#(d)
dowfirst10percent = irfft(first10percent)
plot(dow)
plot(dowfirst10percent)
show()
#(e)
first2percent = zeros(N, float)
first2percent[0 : int(N / 50)] = copy(dowFourier[0 : int(N / 50)])
dowfirst2percent =irfft(first2percent)
plot(dow)
plot(dowfirst2percent)
show()
```
