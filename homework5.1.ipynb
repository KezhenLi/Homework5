```
import urllib.request
from numpy.fft import rfft
from numpy import loadtxt
from matplotlib.pylab import plot, show
```

```
#(a)
url1='http://www-personal.umich.edu/~mejn/cp/data/piano.txt'
with urllib.request.urlopen(url1) as response:
     html = response.read()
data1=html.decode().split("\n")
piano = loadtxt(data1, float)
plot(piano)
show()
pianoFourier = rfft(piano)
plot(abs(pianoFourier[0:9999]))
show()

url2='http://www-personal.umich.edu/~mejn/cp/data/trumpet.txt'
with urllib.request.urlopen(url2) as response:
     html = response.read()
data2=html.decode().split("\n")
trumpet = loadtxt(data2, float)
plot(trumpet)
show()
trumpetFourier = rfft(trumpet)
plot(abs(trumpetFourier[0:9999]))
show()

#(b)
# f =  k * (44100 samples/s) / 100000 samples, k = 2000, f = 882Hz, they're playing musical note A5.
```
